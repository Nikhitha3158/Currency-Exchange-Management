```python
def mean_absolute_percentage_error(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.impute import SimpleImputer

# Assuming aquatss dataframe and preprocessing steps are already defined

# Create a list of n_estimators values to test
n_estimators_values = [600]

# Initialize lists to store evaluation metrics for each n_estimators value
r2_scores = []
mae_scores = []
rmse_scores = []
bias_values = []

for n_estimators in n_estimators_values:
    # Create the model with current n_estimators value
    rf_model = RandomForestRegressor(n_estimators=n_estimators, random_state=39)

    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=39)

    # Impute missing values in X_train and X_test
    imputer = SimpleImputer(strategy='mean')
    X_train_imputed = imputer.fit_transform(X_train)
    X_test_imputed = imputer.transform(X_test)

    # Fit the model with imputed data
    rf_model.fit(X_train_imputed, y_train)
    test_predictions = rf_model.predict(X_test_imputed)

    # Denormalize the predictions
    test_predictions_denormalized = np.exp(test_predictions * sd1 + m1)
    y_test_denormalized = np.exp(y_test * sd1 + m1)

    # Calculate evaluation metrics
    r2 = r2_score(y_test_denormalized, test_predictions_denormalized)
    rmse = np.sqrt(mean_squared_error(y_test_denormalized, test_predictions_denormalized))
    mae = mean_absolute_error(y_test_denormalized, test_predictions_denormalized)
    bias = np.mean(test_predictions_denormalized - y_test_denormalized)

    # Append the metrics to the lists
    r2_scores.append(r2)
    rmse_scores.append(rmse)
    mae_scores.append(mae)
    bias_values.append(bias)

# Print the evaluation metrics for each n_estimators value
for i, n_estimators in enumerate(n_estimators_values):
    print(f"n_estimators = {n_estimators}:")
    print("  R2 Score:", r2_scores[i])
    print("  MAE:", mae_scores[i])
    print("  RMSE:", rmse_scores[i])
    print("  Bias:", bias_values[i])
    print()
```
